# DAT203M — Student Stress Level Prediction

---

## Section 1 — Introduction

### 1.1 Problem Statement

Student mental health is an increasingly critical concern in higher education. Stress affects academic performance, physical health, and long-term well-being. Understanding which lifestyle, financial, and academic factors best predict stress levels enables institutions to design targeted support systems.

This project tackles a **multi-class classification task**:

> **"Predict student stress levels — binned into Low, Average, or High — based on lifestyle, financial, social, and academic survey features."

The target variable `stress` (0–10 scale) is binned into:
- **Class 0 — Low Stress**: score in [0, 3]
- **Class 1 — Average Stress**: score in (3, 6]
- **Class 2 — High Stress**: score in (6, 10]

Three machine learning models are trained and compared:
1. **Softmax Regression** — linear multi-class classifier trained with gradient descent. Interpretable baseline.
2. **k-Nearest Neighbours (k-NN)** — non-parametric distance-based classifier. No gradient descent. Fundamentally different family.
3. **Neural Network (MLP)** — feedforward network with hidden layers. Required by spec. Gradient-descent trained.

### 1.2 Why This Task Is Interesting

Stress is pervasive but often invisible. A predictive model built on easy-to-collect lifestyle data serves as an early-warning tool — flagging high-risk students before they reach crisis. The dataset's breadth (finances, relationships, substance use, academics) enables a holistic view of student well-being that neither a purely academic nor purely social model could achieve.

### 1.3 Dataset Source

The dataset `data1001_survey_data_2025_S2-1.csv` is a real-world anonymous survey from University of Sydney students in DATA1001/DATA1901 across three cohorts: 2024S2, 2025S1, 2025S2.

## Section 4 — Data Cleaning, Preprocessing, and EDA

All data preparation lives in `DataPipeline`. Every model calls `pipeline.get_clean_dataset()` and `pipeline.get_X_y()` — no model ever touches raw data.

```python
class DataPipeline:
    """
    Handles all data loading, cleaning, feature engineering, and splitting.
    A single shared instance is passed to every model.
    """
    NUMERICAL_FEATURES = [
        'hours_work','social_media_use','rent','friends_count',
        'dates','standard_drinks','countries','semesters',
        'commute','data_interest','mark_goal','hours_studying'
    ]
    CATEGORICAL_FEATURES = [
        'relationship_status','drug_use_ans','student_type',
        'lecture_mode','study_type','learner_style'
    ]
    TARGET = 'stress'

    def __init__(self, filepath):
        self.filepath = filepath
        self.df_raw = self.df_clean = None
        self.label_encoders = {}
        self.feature_names  = []
        self._loaded = False

    def prepare(self):
        """Run full pipeline: load → filter → clean → encode → bin."""
        self._load()
        self._filter_consent()
        self._handle_missing()
        self._winsorise()
        self._encode_categoricals()
        self._bin_target()
        self._loaded = True
        print(f'Pipeline ready. Clean dataset: {self.df_clean.shape}')
        return self

    def get_clean_dataset(self):
        assert self._loaded, 'Call prepare() first.'
        return self.df_clean.copy()

    def get_X_y(self):
        assert self._loaded
        enc_cols = [c + '_enc' for c in self.CATEGORICAL_FEATURES]
        all_cols = self.NUMERICAL_FEATURES + enc_cols
        self.feature_names = all_cols
        X = self.df_clean[all_cols].values.astype(float)
        y = self.df_clean['stress_class'].values.astype(int)
        return X, y

    def _load(self):
        self.df_raw = pd.read_csv(self.filepath)
        print(f'Raw shape: {self.df_raw.shape}')

    def _filter_consent(self):
        mask = self.df_raw['consent'] == 'I consent to take part in the study'
        self.df_clean = self.df_raw[mask].copy().reset_index(drop=True)
        feats = self.NUMERICAL_FEATURES + self.CATEGORICAL_FEATURES + [self.TARGET]
        self.df_clean = self.df_clean[feats].dropna(subset=[self.TARGET])
        print(f'After consent filter: {self.df_clean.shape}')

    def _handle_missing(self):
        # Numerical: median (robust to outliers)
        for col in self.NUMERICAL_FEATURES:
            self.df_clean[col].fillna(self.df_clean[col].median(), inplace=True)
        # Categorical: mode
        for col in self.CATEGORICAL_FEATURES:
            self.df_clean[col].fillna(self.df_clean[col].mode()[0], inplace=True)
        print(f'Missing after imputation: {self.df_clean.isnull().sum().sum()}')

    def _winsorise(self):
        """Clip each numerical column to its 1st–99th percentile."""
        for col in self.NUMERICAL_FEATURES:
            lo = np.percentile(self.df_clean[col], 1)
            hi = np.percentile(self.df_clean[col], 99)
            self.df_clean[col] = self.df_clean[col].clip(lo, hi)
        print('Winsorisation complete.')

    def _encode_categoricals(self):
        for col in self.CATEGORICAL_FEATURES:
            le = LabelEncoder()
            self.df_clean[col + '_enc'] = le.fit_transform(
                self.df_clean[col].astype(str))
            self.label_encoders[col] = le
        print('Categorical encoding complete.')

    def _bin_target(self):
        def _bin(v):
            if   v <= 3: return 0
            elif v <= 6: return 1
            else:        return 2
        self.df_clean['stress_class'] = self.df_clean[self.TARGET].apply(_bin)
        counts = self.df_clean['stress_class'].value_counts().sort_index()
        labels = {0:'Low', 1:'Average', 2:'High'}
        for k, v in counts.items():
            print(f'  Class {k} [{labels[k]}]: {v} ({v/len(self.df_clean)*100:.1f}%)')


pipeline = DataPipeline('data1001_survey_data_2025_S2-1.csv')
pipeline.prepare()
```

## Section 5 — Dataset Split

**Ratio**: 70% training / 15% validation / 15% test (stratified).

**Data leakage prevention**: `StandardScaler` fit **only** on training data, then applied identically to validation and test.

```python
X, y = pipeline.get_X_y()

# Step 1: hold out 15% test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=RANDOM_SEED, stratify=y)
# Step 2: split remaining 85% → 70% train, 15% val
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.15/0.85,
    random_state=RANDOM_SEED, stratify=y_temp)

print(f'Training:   {len(y_train):>4} ({len(y_train)/len(y)*100:.1f}%)')
print(f'Validation: {len(y_val):>4} ({len(y_val)/len(y)*100:.1f}%)')
print(f'Test:       {len(y_test):>4} ({len(y_test)/len(y)*100:.1f}%)')

# Scaler fit on training ONLY — prevents data leakage
scaler         = StandardScaler()
X_train_sc     = scaler.fit_transform(X_train)
X_val_sc       = scaler.transform(X_val)
X_test_sc      = scaler.transform(X_test)

print('Scaler fit on training data only.')
visualiser.plot_split_distribution(y_train, y_val, y_test)
```

## Section 6 — Model Training

### Shared utilities

```python
# =====================================================================
# SHARED UTILITIES
# =====================================================================

def softmax(Z):
    Z_s = Z - np.max(Z, axis=1, keepdims=True)
    e   = np.exp(Z_s)
    return e / e.sum(axis=1, keepdims=True)

def relu(Z):       return np.maximum(0, Z)
def relu_grad(Z):  return (Z > 0).astype(float)

def one_hot(y, n=3):
    Y = np.zeros((len(y), n))
    Y[np.arange(len(y)), y] = 1
    return Y

def compute_class_weights(y):
    cls, cnt = np.unique(y, return_counts=True)
    w = len(y) / (len(cls) * cnt)
    return dict(zip(cls, w))

def weighted_xent(Y_pred, Y_true, cw, y_lbl):
    eps = 1e-9
    sw  = np.array([cw[l] for l in y_lbl])
    lp  = -np.sum(Y_true * np.log(Y_pred + eps), axis=1)
    return np.sum(sw * lp) / sw.sum()

CW          = compute_class_weights(y_train)
Y_TR_OH     = one_hot(y_train)
Y_VAL_OH    = one_hot(y_val)
Y_TEST_OH   = one_hot(y_test)
N_FEAT      = X_train_sc.shape[1]

print('Class weights:', {k: round(v,3) for k,v in CW.items()})
```

### Model 1 — Softmax Regression

**Justification**: Multi-class extension of logistic regression. A linear model trained end-to-end with mini-batch gradient descent and cross-entropy loss. Serves as the interpretable baseline — establishes a performance floor for the more complex models. If the neural network cannot beat it, the architecture or training is flawed.

**Optimizer**: Mini-batch SGD + momentum (momentum=0.9)  
**LR Schedule**: Step decay — lr × 0.95 every 10 epochs

```python
class SoftmaxRegression:
    """
    Multi-class softmax regression.
    Trained with mini-batch SGD + momentum + step-decay LR.
    Regularisation: L2 weight penalty.
    """

    def __init__(self, lr=0.01, lr_decay=0.95, decay_every=10,
                 l2=0.001, batch_size=64, n_epochs=100):
        self.lr=lr; self.lr_decay=lr_decay; self.decay_every=decay_every
        self.l2=l2; self.batch_size=batch_size; self.n_epochs=n_epochs
        self.W=self.b=self.vW=self.vb=None
        self.train_losses=[]; self.val_losses=[]
        self.train_accs=[];   self.val_accs=[]
        self._Xv=self._yv=None

    class _Visualiser:
        def __init__(self, o): self._m=o; self._v=Visualisation()
        def plotting(self):
            self._v.plot_learning_curves(
                self._m.train_losses, self._m.val_losses,
                self._m.train_accs,   self._m.val_accs, 'Softmax Regression')
        def correlation(self):
            self._v.plot_confusion_matrix(
                self._m._yv, self._m.predict(self._m._Xv), 'Softmax Regression')

    def clean(self, Xt, yt, Xv, yv):
        self._Xt=Xt; self._yt=yt; self._Xv=Xv; self._yv=yv

    def data_engineering(self, n_feat, n_cls=3):
        """Xavier weight initialisation."""
        self.W  = np.random.randn(n_feat, n_cls) * np.sqrt(2.0/n_feat)
        self.b  = np.zeros((1, n_cls))
        self.vW = np.zeros_like(self.W)
        self.vb = np.zeros_like(self.b)

    def train(self, Xt, Yoh, yt, Xv, Yv, yv, cw, mom=0.9):
        n=len(Xt); lr=self.lr
        for ep in range(self.n_epochs):
            if (ep+1) % self.decay_every == 0: lr *= self.lr_decay
            idx = np.random.permutation(n)
            Xs,Ys,ys = Xt[idx],Yoh[idx],yt[idx]
            for s in range(0, n, self.batch_size):
                Xb=Xs[s:s+self.batch_size]; Yb=Ys[s:s+self.batch_size]
                yb=ys[s:s+self.batch_size]
                sw=np.array([cw[l] for l in yb]); sw/=sw.sum()
                P  = softmax(Xb@self.W+self.b)
                dZ = (P-Yb)*sw[:,None]
                dW = Xb.T@dZ + self.l2*self.W
                db = dZ.sum(0, keepdims=True)
                self.vW=mom*self.vW-lr*dW; self.vb=mom*self.vb-lr*db
                self.W+=self.vW; self.b+=self.vb
            Pt=softmax(Xt@self.W+self.b); Pv=softmax(Xv@self.W+self.b)
            self.train_losses.append(weighted_xent(Pt,Yoh,cw,yt))
            self.val_losses.append(  weighted_xent(Pv,Yv, cw,yv))
            self.train_accs.append(accuracy_score(yt,np.argmax(Pt,1)))
            self.val_accs.append(  accuracy_score(yv,np.argmax(Pv,1)))

    def predict(self, X):       return np.argmax(softmax(X@self.W+self.b), axis=1)
    def predict_proba(self, X): return softmax(X@self.W+self.b)

    def visualise_model(self):
        viz=self._Visualiser(self); viz.plotting(); viz.correlation()

print('SoftmaxRegression defined.')
```

#### Hyperparameter Tuning — Softmax Regression

Grid search: `lr`, `l2_lambda`, `batch_size` — best selected by validation accuracy.

```python
sr_grid = {'lr':[0.05,0.01,0.005], 'l2':[0.0001,0.001,0.01], 'batch_size':[32,64,128]}
best_sr_acc=-1; best_sr_cfg=None; best_sr=None; sr_results=[]

for cfg in ParameterGrid(sr_grid):
    np.random.seed(RANDOM_SEED)
    m = SoftmaxRegression(lr=cfg['lr'], l2=cfg['l2'],
                          batch_size=cfg['batch_size'], n_epochs=80)
    m.clean(X_train_sc, y_train, X_val_sc, y_val)
    m.data_engineering(N_FEAT)
    m.train(X_train_sc, Y_TR_OH, y_train, X_val_sc, Y_VAL_OH, y_val, CW)
    acc = m.val_accs[-1]
    sr_results.append({**cfg, 'val_acc': round(acc,4)})
    if acc > best_sr_acc:
        best_sr_acc=acc; best_sr_cfg=cfg; best_sr=m

print(f'Best config: {best_sr_cfg} | Val acc: {best_sr_acc:.4f}')
pd.DataFrame(sr_results).sort_values('val_acc', ascending=False).head(10)
```

```python
# Learning curves and confusion matrix via visualise_model()
best_sr.visualise_model()
```

### Model 2 — k-Nearest Neighbours (k-NN)

**Justification**: k-NN is a fundamentally different model family from both gradient-descent and tree-based models. It is non-parametric — no explicit training, no gradient descent, no assumptions about functional form. Classification is by majority vote among the k closest training samples using Euclidean distance. This naturally captures local non-linear structure in feature space.

**Why suitable here**: After standardisation, clusters of students with similar financial, social, and academic profiles may share stress levels. k-NN is ideal for capturing such local neighbourhood patterns without imposing a global model structure.

**No gradient descent is used.** This makes k-NN a genuinely distinct methodological choice satisfying the 'different kinds of model' requirement.

**Hyperparameter**: k — the number of neighbours. Tuned by exhaustive search over k = 1..31 (odd values only, to avoid ties).

```python
class KNearestNeighbours:
    """
    k-Nearest Neighbours classifier (from scratch).
    Prediction: majority vote among k closest training samples.
    Distance metric: Euclidean (vectorised via matrix ops).
    No gradient descent — entirely distance-based.
    """

    def __init__(self):
        self.k=None
        self._Xt=self._yt=self._Xv=self._yv=None
        self._k_vals=[]; self._k_accs=[]
        self.visualizer = Visualisation()

    class _Visualiser:
        def __init__(self, o): self._m=o; self._v=Visualisation()
        def plotting(self):
            self._v.plot_knn_k_search(self._m._k_vals, self._m._k_accs)
        def correlation(self):
            self._v.plot_confusion_matrix(
                self._m._yv, self._m.predict(self._m._Xv),
                f'k-NN (k={self._m.k})')

    def clean(self, Xt, yt, Xv, yv):
        self._Xt=Xt; self._yt=yt; self._Xv=Xv; self._yv=yv

    def data_engineering(self):
        """k-NN has no weights to initialise — stored dataset IS the model."""
        pass

    def train(self, k_range=None):
        """
        'Training' k-NN means selecting optimal k by validation accuracy.
        The model memorises (X_train, y_train); no gradient computation needed.
        """
        if k_range is None:
            k_range = list(range(1, 32, 2))
        self._k_vals = k_range
        self._k_accs = []
        for k in k_range:
            preds = self._predict_with_k(self._Xv, k)
            self._k_accs.append(accuracy_score(self._yv, preds))
        best_idx = int(np.argmax(self._k_accs))
        self.k   = k_range[best_idx]
        print(f'Best k = {self.k} | Val accuracy = {self._k_accs[best_idx]:.4f}')

    def _euclidean_dist(self, X_test):
        """Vectorised pairwise Euclidean distance."""
        sq_tr = np.sum(self._Xt**2, axis=1)           # (n_tr,)
        sq_te = np.sum(X_test**2, axis=1, keepdims=True) # (n_te,1)
        dot   = X_test @ self._Xt.T                   # (n_te, n_tr)
        return np.sqrt(np.maximum(sq_te + sq_tr - 2*dot, 0))

    def _predict_with_k(self, X_test, k):
        dists  = self._euclidean_dist(X_test)
        k_idx  = np.argsort(dists, axis=1)[:, :k]
        k_lbls = self._yt[k_idx]
        return np.array([
            np.bincount(row, minlength=3).argmax() for row in k_lbls])

    def predict(self, X):
        assert self.k is not None, 'Call train() first.'
        return self._predict_with_k(X, self.k)

    def visualise_model(self):
        viz=self._Visualiser(self); viz.plotting(); viz.correlation()

print('KNearestNeighbours defined.')
```

```python
knn = KNearestNeighbours()
knn.clean(X_train_sc, y_train, X_val_sc, y_val)
knn.data_engineering()
knn.train(k_range=list(range(1, 32, 2)))
knn.visualise_model()

print(f'\nFinal hyperparameter: k = {knn.k}')
print('Low k = overfitting (memorises noise); high k = underfitting (over-smoothed boundary).')
print('The search plot shows the bias-variance tradeoff across k values.')
```

### Model 3 — Neural Network (Multi-Layer Perceptron)

**Justification**: Required by the project specification. The MLP extends softmax regression by adding hidden layers with non-linear ReLU activations, enabling it to learn complex feature interactions that no linear model can capture. Backpropagation computes exact gradients through all layers; mini-batch SGD with momentum performs the weight updates.

**Architecture**: Input → Dense(ReLU) → Dropout → Output(Softmax)  
**Optimizer**: SGD with momentum (momentum=0.9)  
**LR Schedule**: Step decay — lr × 0.95 every 10 epochs  
**Regularisation**: Inverted dropout + L2 weight penalty

```python
class NeuralNetwork:
    """
    Feedforward MLP (1 hidden layer).
    Input → Hidden (ReLU + Dropout) → Output (Softmax).
    Trained with mini-batch SGD + momentum + step-decay LR.
    """

    def __init__(self, n_hidden=64, lr=0.01, lr_decay=0.95,
                 decay_every=10, l2=0.001, dropout=0.2,
                 mom=0.9, batch_size=64, n_epochs=150):
        self.n_hidden=n_hidden; self.lr=lr; self.lr_decay=lr_decay
        self.decay_every=decay_every; self.l2=l2; self.dropout=dropout
        self.mom=mom; self.batch_size=batch_size; self.n_epochs=n_epochs
        self.W1=self.b1=self.W2=self.b2=None
        self.vW1=self.vb1=self.vW2=self.vb2=None
        self.train_losses=[]; self.val_losses=[]
        self.train_accs=[];   self.val_accs=[]
        self._Xv=self._yv=None

    class _Visualiser:
        def __init__(self, o): self._m=o; self._v=Visualisation()
        def plotting(self):
            self._v.plot_learning_curves(
                self._m.train_losses, self._m.val_losses,
                self._m.train_accs,   self._m.val_accs, 'Neural Network (MLP)')
        def correlation(self):
            self._v.plot_confusion_matrix(
                self._m._yv, self._m.predict(self._m._Xv), 'Neural Network (MLP)')

    def clean(self, Xt, yt, Xv, yv):
        self._Xt=Xt; self._yt=yt; self._Xv=Xv; self._yv=yv

    def data_engineering(self, n_feat, n_cls=3):
        """Xavier / He weight initialisation."""
        self.W1=np.random.randn(n_feat, self.n_hidden)*np.sqrt(2.0/n_feat)
        self.b1=np.zeros((1,self.n_hidden))
        self.W2=np.random.randn(self.n_hidden, n_cls)*np.sqrt(2.0/self.n_hidden)
        self.b2=np.zeros((1,n_cls))
        self.vW1=np.zeros_like(self.W1); self.vb1=np.zeros_like(self.b1)
        self.vW2=np.zeros_like(self.W2); self.vb2=np.zeros_like(self.b2)

    def _fwd(self, X, training=False):
        Z1=X@self.W1+self.b1; A1=relu(Z1)
        if training and self.dropout>0:
            mask=(np.random.rand(*A1.shape)>self.dropout).astype(float)
            A1=A1*mask/(1-self.dropout); self._mask=mask
        else: self._mask=None
        Z2=A1@self.W2+self.b2; A2=softmax(Z2)
        self._cache=(X,Z1,A1,Z2,A2)
        return A2

    def _bwd(self, Yt, sw):
        X,Z1,A1,Z2,A2=self._cache; sw=sw/sw.sum()
        dZ2=(A2-Yt)*sw[:,None]
        dW2=A1.T@dZ2+self.l2*self.W2; db2=dZ2.sum(0,keepdims=True)
        dA1=dZ2@self.W2.T
        if self._mask is not None: dA1=dA1*self._mask/(1-self.dropout)
        dZ1=dA1*relu_grad(Z1)
        dW1=X.T@dZ1+self.l2*self.W1; db1=dZ1.sum(0,keepdims=True)
        return dW1,db1,dW2,db2

    def train(self, Xt, Yoh, yt, Xv, Yv, yv, cw):
        n=len(Xt); lr=self.lr
        for ep in range(self.n_epochs):
            if (ep+1)%self.decay_every==0: lr*=self.lr_decay
            idx=np.random.permutation(n)
            Xs,Ys,ys=Xt[idx],Yoh[idx],yt[idx]
            for s in range(0,n,self.batch_size):
                Xb=Xs[s:s+self.batch_size]; Yb=Ys[s:s+self.batch_size]
                yb=ys[s:s+self.batch_size]
                sw=np.array([cw[l] for l in yb])
                self._fwd(Xb, training=True)
                dW1,db1,dW2,db2=self._bwd(Yb,sw)
                self.vW1=self.mom*self.vW1-lr*dW1; self.vb1=self.mom*self.vb1-lr*db1
                self.vW2=self.mom*self.vW2-lr*dW2; self.vb2=self.mom*self.vb2-lr*db2
                self.W1+=self.vW1; self.b1+=self.vb1
                self.W2+=self.vW2; self.b2+=self.vb2
            Pt=self._fwd(Xt); Pv=self._fwd(Xv)
            self.train_losses.append(weighted_xent(Pt,Yoh,cw,yt))
            self.val_losses.append(  weighted_xent(Pv,Yv, cw,yv))
            self.train_accs.append(accuracy_score(yt,np.argmax(Pt,1)))
            self.val_accs.append(  accuracy_score(yv,np.argmax(Pv,1)))

    def predict(self, X):       return np.argmax(self._fwd(X), axis=1)
    def predict_proba(self, X): return self._fwd(X)

    def visualise_model(self):
        viz=self._Visualiser(self); viz.plotting(); viz.correlation()

print('NeuralNetwork defined.')
```

#### Hyperparameter Tuning — Neural Network

Grid search: `lr`, `n_hidden`, `dropout`, `l2`. Best by validation accuracy.

```python
nn_grid = {
    'lr':[0.05,0.01,0.005], 'n_hidden':[32,64,128],
    'dropout':[0.1,0.2,0.3], 'l2':[0.0001,0.001]
}
best_nn_acc=-1; best_nn_cfg=None; best_nn=None; nn_results=[]

for cfg in ParameterGrid(nn_grid):
    np.random.seed(RANDOM_SEED)
    m = NeuralNetwork(n_hidden=cfg['n_hidden'], lr=cfg['lr'],
                      l2=cfg['l2'], dropout=cfg['dropout'], n_epochs=80)
    m.clean(X_train_sc, y_train, X_val_sc, y_val)
    m.data_engineering(N_FEAT)
    m.train(X_train_sc, Y_TR_OH, y_train, X_val_sc, Y_VAL_OH, y_val, CW)
    acc=m.val_accs[-1]
    nn_results.append({**cfg, 'val_acc':round(acc,4)})
    if acc>best_nn_acc:
        best_nn_acc=acc; best_nn_cfg=cfg; best_nn=m

print(f'Best config: {best_nn_cfg} | Val acc: {best_nn_acc:.4f}')
pd.DataFrame(nn_results).sort_values('val_acc', ascending=False).head(10)
```

## Section 8 — Improving Model Performance

**Error analysis findings**: 'Average Stress' shows the lowest F1 across all models due to central boundary overlap. Neural Network is the best baseline model.

**Improvement strategies applied to Neural Network (best model):**
1. **Data-level**: Feature engineering — pairwise interaction terms between top-5 stress-correlated features. Gives the network richer non-linear signal without changing the architecture.
2. **Model-level**: Second hidden layer (deeper architecture) + dropout rate increased to 0.3.

**Why we improve the Neural Network and not the others**: Softmax is a linear model by design; adding interactions helps but the fundamental limit is linearity. k-NN already uses all pairwise distances and does not benefit from explicit interactions the same way. The Neural Network has the capacity to exploit richer features and a deeper architecture.

```python
# Feature engineering: pairwise interactions between top-5 correlated features
corr_s = df_clean[pipeline.NUMERICAL_FEATURES+[pipeline.TARGET]]\
    .corr()[pipeline.TARGET].drop(pipeline.TARGET).abs()
top5 = corr_s.nlargest(5).index.tolist()
print('Top 5 features for interactions:', top5)

all_feat = pipeline.feature_names

def add_interactions(X_raw, top_feats, feat_cols):
    idxs   = [feat_cols.index(f) for f in top_feats if f in feat_cols]
    extras = [(X_raw[:,i]*X_raw[:,j]).reshape(-1,1)
               for i in range(len(idxs)) for j in range(i+1,len(idxs))]
    return np.hstack([X_raw]+extras) if extras else X_raw

X_tr_aug  = add_interactions(X_train, top5, all_feat)
X_val_aug = add_interactions(X_val,   top5, all_feat)
X_te_aug  = add_interactions(X_test,  top5, all_feat)

sc2=StandardScaler()
X_tr_aug_sc  = sc2.fit_transform(X_tr_aug)
X_val_aug_sc = sc2.transform(X_val_aug)
X_te_aug_sc  = sc2.transform(X_te_aug)

print(f'Feature dim: {X_train.shape[1]} → {X_tr_aug.shape[1]}')
```

```python
class DeepNeuralNetwork:
    """
    3-layer MLP (2 hidden layers).
    Input → Hidden1 (ReLU+Dropout) → Hidden2 (ReLU+Dropout) → Output (Softmax)
    Optimizer: SGD + momentum. Step-decay LR. L2 + Dropout regularisation.
    """

    def __init__(self, n_h1=64, n_h2=32, lr=0.01, lr_decay=0.95,
                 decay_every=10, l2=0.001, dropout=0.3,
                 mom=0.9, batch_size=64, n_epochs=150):
        self.n_h1=n_h1; self.n_h2=n_h2; self.lr=lr
        self.lr_decay=lr_decay; self.decay_every=decay_every
        self.l2=l2; self.dropout=dropout; self.mom=mom
        self.batch_size=batch_size; self.n_epochs=n_epochs
        self.train_losses=[]; self.val_losses=[]
        self.train_accs=[];   self.val_accs=[]
        self._Xv=self._yv=None

    class _Visualiser:
        def __init__(self, o): self._m=o; self._v=Visualisation()
        def plotting(self):
            self._v.plot_learning_curves(
                self._m.train_losses, self._m.val_losses,
                self._m.train_accs,   self._m.val_accs,
                'Deep Neural Network (improved)')
        def correlation(self):
            self._v.plot_confusion_matrix(
                self._m._yv, self._m.predict(self._m._Xv),
                'Deep Neural Network (improved)')

    def clean(self, Xt, yt, Xv, yv):
        self._Xt=Xt; self._yt=yt; self._Xv=Xv; self._yv=yv

    def data_engineering(self, n_feat, n_cls=3):
        self.W1=np.random.randn(n_feat,self.n_h1)*np.sqrt(2.0/n_feat)
        self.b1=np.zeros((1,self.n_h1))
        self.W2=np.random.randn(self.n_h1,self.n_h2)*np.sqrt(2.0/self.n_h1)
        self.b2=np.zeros((1,self.n_h2))
        self.W3=np.random.randn(self.n_h2,n_cls)*np.sqrt(2.0/self.n_h2)
        self.b3=np.zeros((1,n_cls))
        for a in ['vW1','vb1','vW2','vb2','vW3','vb3']:
            setattr(self, a, np.zeros_like(getattr(self, a.replace('v',''))))

    def _drop(self, A, training):
        if training and self.dropout>0:
            m=(np.random.rand(*A.shape)>self.dropout).astype(float)
            return A*m/(1-self.dropout), m
        return A, None

    def _fwd(self, X, training=False):
        Z1=X@self.W1+self.b1;  A1=relu(Z1); A1,m1=self._drop(A1,training)
        Z2=A1@self.W2+self.b2; A2=relu(Z2); A2,m2=self._drop(A2,training)
        Z3=A2@self.W3+self.b3; A3=softmax(Z3)
        self._cache=(X,Z1,A1,m1,Z2,A2,m2,Z3,A3)
        return A3

    def _bwd(self, Yt, sw):
        X,Z1,A1,m1,Z2,A2,m2,Z3,A3=self._cache; sw=sw/sw.sum()
        dZ3=(A3-Yt)*sw[:,None]
        dW3=A2.T@dZ3+self.l2*self.W3; db3=dZ3.sum(0,keepdims=True)
        dA2=dZ3@self.W3.T
        if m2 is not None: dA2=dA2*m2/(1-self.dropout)
        dZ2=dA2*relu_grad(Z2)
        dW2=A1.T@dZ2+self.l2*self.W2; db2=dZ2.sum(0,keepdims=True)
        dA1=dZ2@self.W2.T
        if m1 is not None: dA1=dA1*m1/(1-self.dropout)
        dZ1=dA1*relu_grad(Z1)
        dW1=X.T@dZ1+self.l2*self.W1; db1=dZ1.sum(0,keepdims=True)
        return dW1,db1,dW2,db2,dW3,db3

    def train(self, Xt, Yoh, yt, Xv, Yv, yv, cw):
        n=len(Xt); lr=self.lr
        for ep in range(self.n_epochs):
            if (ep+1)%self.decay_every==0: lr*=self.lr_decay
            idx=np.random.permutation(n)
            Xs,Ys,ys=Xt[idx],Yoh[idx],yt[idx]
            for s in range(0,n,self.batch_size):
                Xb=Xs[s:s+self.batch_size]; Yb=Ys[s:s+self.batch_size]
                yb=ys[s:s+self.batch_size]
                sw=np.array([cw[l] for l in yb])
                self._fwd(Xb,True)
                gs=self._bwd(Yb,sw)
                for p,v,g in zip(['W1','b1','W2','b2','W3','b3'],
                                  ['vW1','vb1','vW2','vb2','vW3','vb3'],gs):
                    nv=self.mom*getattr(self,v)-lr*g
                    setattr(self,v,nv); setattr(self,p,getattr(self,p)+nv)
            Pt=self._fwd(Xt); Pv=self._fwd(Xv)
            self.train_losses.append(weighted_xent(Pt,Yoh,cw,yt))
            self.val_losses.append(  weighted_xent(Pv,Yv, cw,yv))
            self.train_accs.append(accuracy_score(yt,np.argmax(Pt,1)))
            self.val_accs.append(  accuracy_score(yv,np.argmax(Pv,1)))

    def predict(self, X): return np.argmax(self._fwd(X), axis=1)

    def visualise_model(self):
        viz=self._Visualiser(self); viz.plotting(); viz.correlation()

print('DeepNeuralNetwork defined.')
```

```python
np.random.seed(RANDOM_SEED)
n_aug = X_tr_aug_sc.shape[1]

deep_nn = DeepNeuralNetwork(
    n_h1=best_nn_cfg.get('n_hidden',64),
    n_h2=max(16, best_nn_cfg.get('n_hidden',64)//2),
    lr=best_nn_cfg.get('lr',0.01),
    l2=best_nn_cfg.get('l2',0.001),
    dropout=0.3, n_epochs=150
)
deep_nn.clean(X_tr_aug_sc, y_train, X_val_aug_sc, y_val)
deep_nn.data_engineering(n_aug)
deep_nn.train(X_tr_aug_sc, Y_TR_OH, y_train,
              X_val_aug_sc, Y_VAL_OH, y_val, CW)
print(f'Improved model val accuracy: {deep_nn.val_accs[-1]:.4f}')
```

### 9.1 Discussion

**Softmax Regression** is the linear interpretable baseline. It uses gradient descent on a single weight matrix — the simplest possible gradient-descent model. Any model below its performance has a training or architecture problem.

**k-NN** uses no gradient descent whatsoever. It classifies by proximity in standardised feature space. Its advantage is that it requires no assumptions about the data distribution; its weakness is sensitivity to the choice of k and the curse of dimensionality at high feature counts.

**Neural Network (MLP)** outperforms the linear model by learning non-linear combinations of features through its hidden layer. The combination of dropout + L2 + weighted cross-entropy addresses both overfitting and class imbalance.

**Deep NN (improved)** targets the 'Average Stress' misclassification identified in error analysis. Feature interactions give the model explicit access to combined signals (e.g. high rent AND high social media use AND complicated relationship), while the second hidden layer increases representational capacity.

The model ranked highest by **Macro F1** is the recommended best model, since Macro F1 penalises poor performance on minority classes (Low and High stress) equally to the majority class.
